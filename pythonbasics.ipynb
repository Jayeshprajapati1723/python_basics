Collection Datatypes/Sequences: String, List, Tuple, Set, Dict

Mutable/Immutable,Ordered/Unordered(index),Allow Duplicates or not

- String: Immutable,ordered, allow duplicates

- List: Mutable, Ordered, Allow Duplicates

- Tuple: Immutable, Ordered, allow duplicates

- Set: Mutable, Unordered, DO NOT allow duplicates

- Dict: Follows Key-value pairs, unordered, keys-immutable, value-mutable, keys-no duplicates, values-allow duplicates

```python
#String Slicing
#str[start:stop:step]  start:inclusive Stop:exclusive
#default start:0, stop:end of the string, step:1
var="abcdefghijklmnop"
print(var[0:5])
print(var[:5])
print(var[5:])
print(var[0:10:2])
print(var[::-1]) #reverse string
```

```python
import random as rd

sec = rd.randint(1, 100)
guess = 0 # Loop chalane ke liye initial value deni zaroori hai
attempts = 0

print("Welcome to the Guessing Game! (1 to 100)")

while guess != sec:
    guess = int(input("Enter your guess: "))
    attempts += 1 # Har baar guess lene par attempt badhao

    # Ye sara part loop ke andar hona chahiye (Indented)
    if guess == sec:
        print("Congratulations! You guessed it.")
        print("Total attempts:", attempts)
    elif guess > sec:
        print("Your guess is high! Try again.")
    else:
        print("Your guess is low! Try again.")
```
